```
# Most and Least profitable Lead Studio. (Take an average of all Profitability measures grouped by Lead studios)

df_prof = df.groupby("Lead Studio")["Profitability"].mean()
print("Most profitable:")
print(df_prof.sort_values().tail(1))

print("\nLeast profitable:")
print(df_prof.sort_values().head(1))
```

```
# Count of least Frequent Genre with Rotten Tomato score more than 50%.
rotten_50 = df[df["Rotten Tomatoes %"] > 50]
rotten_50_genre = rotten_50.groupby("Genre").count()
rotten_50_sorted = rotten_50_genre.sort_values(by="Rotten Tomatoes %").head(1)
rotten_50_sorted["Film"]
```

```
# Ranking of Fox studio for Comedy movies based on Audience Score. v1
comedy = df[df["Genre"]=="Comedy"]
comedy_leadstudio = comedy.groupby("Lead Studio").mean()
comedy_leadstudio_rank = comedy_leadstudio.rank(ascending= False)["Audience score %"]
comedy_leadstudio_rank["Fox"]
```

```
# Ranking of Fox studio for Comedy movies based on Audience Score. v2
comedy_mov = df[df["Genre"] == "Comedy"]
df_AudienceScore_ave_Studio = pd.DataFrame(comedy_mov.groupby('Lead Studio')['Audience score %'].mean())
df_AudienceScore_ave_Studio['Audience score %'].rank(ascending=False) ["Fox"]
```
